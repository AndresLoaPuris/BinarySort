# Binary Order

```python
import time
```

## Function Read DataSet

```python
def DataSet(data_name, values):
    data = []
    quantity = 0
    with open(data_name) as f:
        
        for line in f:
            data.append([int(x) for x in line.split()])
            
        for i in range(len(data)):
            for j in range(len(data[i])):
                values.append(data[i][j])
                quantity += 1
            
    return values, quantity
```

## Entity Node

```python
class Node():
    def __init__(self, value):
        self.left = None
        self.right = None
        self.data = value
        self.times = 1
```

## Entity Binary Tree

```python
class binaryTree():
    
    def __init__(self):
        self.root = None
        
    def insert(self, node, value):
        if node == None:
            node = Node(value)
        else:
            if value == node.data:
                node.times += 1
            elif value < node.data:
                node.left = self.insert(node.left, value)
            else:
                node.right = self.insert(node.right, value)
                
        return node
    

    def inOrder(self, node , orderedValues):
        if node == None:
            return None
        else:
            self.inOrder(node.left, orderedValues)
            
            if node.times > 1 :
                for _ in range(node.times):
                    orderedValues.append(node.data)       
            else:    
                orderedValues.append(node.data)
            
            self.inOrder(node.right, orderedValues)
            
        return orderedValues
```

## Call Function : Main

```python
randomNumbers , quantity = DataSet("RandomData.txt", [])

binaryOrder = binaryTree()

for number in randomNumbers:
    binaryOrder.root = binaryOrder.insert(binaryOrder.root, number)

start = time.perf_counter()
print(binaryOrder.inOrder(binaryOrder.root, []))
print(" \n -> Order ", quantity, " numbers in ", time.perf_counter() - start, "seg.")
```
